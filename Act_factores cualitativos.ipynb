```python
df = pd.read_excel("Motor Trend Car Road Tests.xlsx")
df.head()
```

```python
df = df.dropna().drop_duplicates()

x = df.drop(["model", "mpg"], axis=1)

y = df["mpg"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.4, random_state=137)
```

```python
scaler = StandardScaler().fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train_scaled, y_train)

y_train_pred = lr.predict(x_train_scaled)
y_test_pred = lr.predict(x_test_scaled)
```

```python
r2_score(y_test, y_test_pred) , r2_score(y_train, y_train_pred)
```
